Sea $n=77021627$, calcula $a^{n-1}$ mod $n$ para las bases $a=2,3,5,7,11$

```python
n=77021627
```

Implementamos un método para calcular la potencia anterior mediante el algoritmo de exponenciación rápida

```python
def exp_rapida(base,exponente,n):
    acu=1
    ex=0
    paso=0
    num=exponente.digits(2)
    num.reverse()
    for i in num:
        paso=paso+1
        acu=(acu*acu)%n
        ex=(2*ex)%n
        
        if i%2==1:
            acu=(acu*base)%n
            ex=(ex+1)%n
        print(base,'^',ex,'mod',n,'=',acu)
    if(acu+1==n):
        acu=-1
    return (acu)
```

```python
base=[2,3,5,7,11]
```

```python
for i in base:
    exp_rapida(i,n-1,n)
    print()
```
