Apartado 2.-Calcula la sucesi칩n de Lucas asociada

```python
def suc_lucas(n,p,q):
    r=n-kronecker(5,n)
    num=r.digits(2)
    num.reverse()
    i=0
    u_anterior=0
    u=1
    aux=0



    #print 'U[%d]=%d V[%d]=%d' %(i,u_anterior,i+1,u)
    #print
    for j in num:
        if (j%2==0):
            i=2*i
            aux=u_anterior
            u_anterior=(2*u_anterior*u-p*u_anterior*u_anterior)%n
            u=(u*u-q*aux*aux)%n
        if (j%2)==1:
            i=2*i+1
            aux=u_anterior
            u_anterior=(u*u-q*aux*aux)%n
            u=(p*u*u-2*q*u*aux)%n
        v=(2*u-p*u_anterior)%n
        print('U[',i,']=',u_anterior)
        print('V[',i,']=',v)
        print()
```
