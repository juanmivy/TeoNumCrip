```python
def FCS(n):
    ##El algoritmo y la explicacion de este esta en los apuntes
    ##Calculamos la parte entera de la raiz de n
    parte_entera_raiz=(sqrt(n).n()).floor()
    p=0
    k=1
    i=0
    q=(p+parte_entera_raiz-(p+parte_entera_raiz)%k)/k
    resultado=[q]
    i=i+1
    p_aux=p
    p=q*k-p
    k_aux=k
    k=(n-p*p)/k_aux
    q=(p+parte_entera_raiz-(p+parte_entera_raiz)%k)/k
    resultado=resultado+[q]
    
    while k!=1:
        i=i+1
        p_aux=p
        k_aux=k
        p=q*k-p
        k=(n-p_aux*p_aux)/k_aux+2*q*p_aux-(q*q)*k_aux
        q=(p+parte_entera_raiz-(p+parte_entera_raiz)%k)/k
        resultado=resultado+[q]
    return(resultado)
```
